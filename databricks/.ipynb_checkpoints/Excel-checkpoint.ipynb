```python
%%spark
val sqlContext = new org.apache.spark.sql.SQLContext(sc)

val superStore = sqlContext.read.format("com.crealytics.spark.excel").option("sheetName", "orders").option("useHeader", "true").load("file:///home/churtado/data/superstore.xls")
superStore.createOrReplaceTempView("markets")
```

```python
%%spark -c sql
SHOW TABLES
```

```python
%%spark -c sql -o df_markets --maxrows 10
SELECT * FROM markets
```
